# Exercise 1 

```python
import numpy as np
import pandas as pd

np.random.seed(1429)

n = 50
REP = 1000
BOOTREP = 499
m = 0
s = 1.3
mu = np.exp(m + 0.5 * (s ** 2))
beta = np.sin(mu)

xbar = np.zeros([REP, 1])
bhat = np.zeros([REP, 1])
SE = np.zeros([REP, 1])
trat = np.zeros([REP, 1])
LCLasym = np.zeros([REP, 1])
UCLasym = np.zeros([REP, 1])

for i in range(REP):
    if ((i + 1) % 100) == 0:
        print(i + 1)
    
    X = np.exp(np.random.normal(loc=m, scale=s, size=([n, 1])))
    xbar[i] = np.mean(X)
    bhat[i] = np.sin(xbar[i])
    SE[i] = np.sqrt(np.var(X) / n * (np.cos(xbar[i])) ** 2)
    trat[i] = (bhat[i] - beta) / SE[i]
    
    LCLasym[i] = bhat[i] - 1.96 * SE[i]
    UCLasym[i] = bhat[i] + 1.96 * SE[i]
    
CoverageFreqasym = np.mean((beta > LCLasym) & (beta < UCLasym)) 
print('Coverage freq.(asym): \n', CoverageFreqasym)
```

#### Bias and bias-corrected estimates

```python
#Bootstrap
np.random.seed(1429)
n = 50

X = np.exp(np.random.normal(loc=m, scale=s, size=([n, 1])))

bhat_bs = np.zeros([BOOTREP, 1])
for b in range(BOOTREP):
    bhat_bs[b] = np.sin(np.mean(X[np.random.randint(0, n - 1, n)]))
    
# Bias
bias_bs = np.mean(bhat_bs) - np.sin(np.mean(X))
bias_bs_correction = 2 * np.sin(np.mean(X)) - np.mean(bhat_bs)

print('Bootstrap bias: ', bias_bs)
print('Bootstrap bias-corrected: ', bias_bs_correction)

#Jackknife

bhat_jk = np.zeros([n, 1])

for i in range(n):
    mask = np.ones(n, dtype=bool)
    mask[i] = 0
    bhat_jk[i] = np.sin(np.mean(X[mask]))
    
bias_jk = (n - 1) * (np.mean(bhat_jk) - np.sin(np.mean(X)))
bias_jk_correction = n * np.sin(np.mean(X)) - (n - 1) * np.mean(bhat_jk)

print('Jackknife bias: ', bias_jk)
print('Jackknife bias-corrected: ', bias_jk_correction)

# Confidence interval
```

#### Confidence interval - bootstrap

```python
np.random.seed(1429)

n = 50
REP = 1000
BOOTREP = 499
m = 0
s = 1.3
mu = np.exp(m + 0.5 * (s ** 2))
beta = np.sin(mu)

xbar = np.zeros([REP, 1])
bhat_bs = np.zeros([REP, 1])
SE = np.zeros([REP, 1])
trat = np.zeros([REP, 1])
LCLasym = np.zeros([REP, 1])
UCLasym = np.zeros([REP, 1])

for i in range(REP):
    if ((i + 1) % 100) == 0:
        print(i + 1)
    
    X = np.exp(np.random.normal(loc=m, scale=s, size=([n, 1])))
    xbar[i] = np.mean(X)
    
    bhat_bs_rep = np.zeros([BOOTREP, 1])
    for b in range(BOOTREP):
        bhat_bs_rep[b] = np.sin(np.mean(X[np.random.randint(0, n - 1, n)]))

    # Bias
    bias_bs_correction = 2 * np.sin(np.mean(X)) - np.mean(bhat_bs_rep)
    bhat_bs[i] = bias_bs_correction
    
    SE[i] = np.sqrt(np.var(X) / n * (np.cos(xbar[i])) ** 2)
    trat[i] = (bhat_bs[i] - beta) / SE[i]
    
    LCLasym[i] = bhat_bs[i] - 1.96 * SE[i]
    UCLasym[i] = bhat_bs[i] + 1.96 * SE[i]
    
CoverageFreqasym = np.mean((beta > LCLasym) & (beta < UCLasym)) 
print('Coverage freq.(asym): \n', CoverageFreqasym)
```

#### Confidence interval - jackknife

```python
np.random.seed(1429)

n = 50
REP = 1000
BOOTREP = 499
m = 0
s = 1.3
mu = np.exp(m + 0.5 * (s ** 2))
beta = np.sin(mu)

xbar = np.zeros([REP, 1])
bhat_jk = np.zeros([REP, 1])
SE = np.zeros([REP, 1])
trat = np.zeros([REP, 1])
LCLasym = np.zeros([REP, 1])
UCLasym = np.zeros([REP, 1])

for i in range(REP):
    if ((i + 1) % 100) == 0:
        print(i + 1)
    
    X = np.exp(np.random.normal(loc=m, scale=s, size=([n, 1])))
    xbar[i] = np.mean(X)
    
    bhat_jk_rep = np.zeros([n, 1])

    for b in range(n):
        mask = np.ones(n, dtype=bool)
        mask[b] = 0
        bhat_jk_rep[b] = np.sin(np.mean(X[mask]))

    bias_jk_correction = n * np.sin(np.mean(X)) - (n - 1) * np.mean(bhat_jk_rep)
    bhat_jk[i] = bias_jk_correction
    
    SE[i] = np.sqrt(np.var(X) / n * (np.cos(xbar[i])) ** 2)
    trat[i] = (bhat_jk[i] - beta) / SE[i]
    
    LCLasym[i] = bhat_jk[i] - 1.96 * SE[i]
    UCLasym[i] = bhat_jk[i] + 1.96 * SE[i]
    
CoverageFreqasym = np.mean((beta > LCLasym) & (beta < UCLasym)) 
print('Coverage freq.(asym): \n', CoverageFreqasym)
```

#### CI Jackknife with corrected SE

```python
import numpy as np
import pandas as pd

np.random.seed(1429)

n = 50
REP = 1000
BOOTREP = 499
m = 0
s = 1.3
mu = np.exp(m + 0.5 * (s ** 2))
beta = np.sin(mu)

xbar = np.zeros([REP, 1])
bhat = np.zeros([REP, 1])
SE = np.zeros([REP, 1])
trat = np.zeros([REP, 1])
LCLasym = np.zeros([REP, 1])
UCLasym = np.zeros([REP, 1])

for i in range(REP):
    if ((i + 1) % 100) == 0:
        print(i + 1)
    
    X = np.exp(np.random.normal(loc=m, scale=s, size=([n, 1])))
    xbar[i] = np.mean(X)
    bhat[i] = np.sin(xbar[i])
    
    bhat_jk = np.zeros([n, 1])

    for b in range(n):
        mask = np.ones(n, dtype=bool)
        mask[b] = 0
        bhat_jk[b] = np.sin(np.mean(X[mask]))
    
    SE[i] = np.sqrt((n - 1) / n * np.sum((bhat_jk - np.mean(bhat_jk)) ** 2))
    trat[i] = (bhat[i] - beta) / SE[i]
    
    LCLasym[i] = bhat[i] - 1.96 * SE[i]
    UCLasym[i] = bhat[i] + 1.96 * SE[i]
    
CoverageFreqasym = np.mean((beta > LCLasym) & (beta < UCLasym)) 
print('Coverage freq.(asym): \n', CoverageFreqasym)
```

#### CI Bootstrap with corrected SE

```python
np.random.seed(1429)

n = 50
REP = 1000
BOOTREP = 499
m = 0
s = 1.3
mu = np.exp(m + 0.5 * (s ** 2))
beta = np.sin(mu)

xbar = np.zeros([REP, 1])
bhat = np.zeros([REP, 1])
SE = np.zeros([REP, 1])
trat = np.zeros([REP, 1])
LCLasym = np.zeros([REP, 1])
UCLasym = np.zeros([REP, 1])

for i in range(REP):
    if ((i + 1) % 100) == 0:
        print(i + 1)
    
    X = np.exp(np.random.normal(loc=m, scale=s, size=([n, 1])))
    xbar[i] = np.mean(X)
    bhat[i] = np.sin(xbar[i])
    
    bhat_bs = np.zeros([BOOTREP, 1])

    for b in range(BOOTREP):
        bhat_bs[b] = np.sin(np.mean(X[np.random.randint(0, n - 1, n)]))
    
    SE[i] = np.sqrt((1 / (BOOTREP - 1)) * np.sum((bhat_bs - np.mean(bhat_bs)) ** 2))
    trat[i] = (bhat[i] - beta) / SE[i]
    
    LCLasym[i] = bhat[i] - 1.96 * SE[i]
    UCLasym[i] = bhat[i] + 1.96 * SE[i]
    
CoverageFreqasym = np.mean((beta > LCLasym) & (beta < UCLasym)) 
print('Coverage freq.(asym): \n', CoverageFreqasym)
```

```python
np.random.seed(1429)

n = 50
REP = 1000
BOOTREP = 499
m = 0
s = 1.3
mu = np.exp(m + 0.5 * (s ** 2))
beta = np.sin(mu)

xbar = np.zeros([REP, 1])
bhat = np.zeros([REP, 1])
SE = np.zeros([REP, 1])
trat = np.zeros([BOOTREP, 1])
LCL = np.zeros([REP, 1])
UCL = np.zeros([REP, 1])
LCLasym = np.zeros([REP, 1])
UCLasym = np.zeros([REP, 1])

for i in range(REP):
    if ((i + 1) % 100) == 0:
        print(i + 1)
        
    X = np.exp(np.random.normal(loc = m, scale = s, size = ([n,1])))
    xbar[i] = np.mean(X)
    
    bhat_bs = np.zeros([BOOTREP,1])
    for b in range(BOOTREP):
        X_bs = X[np.random.randint(0, n-1, n)]
        bhat_bs[b] = np.sin(np.mean(X_bs))
        trat[b] = (bhat_bs[b] - beta) / np.std(X_bs)
        
    LCL[i] = np.percentile(trat, q = 97.5, interpolation = 'midpoint')
    UCL[i] = np.percentile(trat, q = 2.5, interpolation = 'midpoint')
        
    bhat[i] = np.mean(bhat_bs)
    #bhat[i] = np.sin(xbar[i])
    SE[i] = np.sqrt(np.var(X) / n * (np.cos(xbar[i])) ** 2)
    #SE[i] = np.sqrt(np.cos(xbar[i]) * np.var(X) / n * (np.cos(xbar[i])))  
    #SE[i] = np.sqrt(np.var(X) * (np.cos(xbar[i]))**2)  # CODE THE SE YOURSELF!!!
    
    LCLasym[i] = bhat[i] - LCL[i] * SE[i]
    UCLasym[i] = bhat[i] - UCL[i] * SE[i]

 

CoverageFreqasym = np.mean((beta > LCLasym) & (beta < UCLasym))
CoverageFreqasym
```
